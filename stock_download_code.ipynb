```python
import pandas_datareader.data as web  
```

```python
# SBIN.NS 
# yahoo finance 
# start : data starting point 
# end : upto what date we want to import 
df = web.DataReader('SBIN.NS','yahoo',start = '2020-01-01', end = '2021-01-01')  
df
```

```python
companies_list = ['SBIN.NS','ICICIBANK.NS','ITC.NS'] 
df = web.DataReader(companies_list,'yahoo', start = '2020-01-01', end = '2021-01-01') 
df
```

```python
temp_df = df.loc[:,(slice(None), 'SBIN.NS')] 
temp_df.columns =  ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']  
temp_df
```

```python
companies_list = ['SBIN.NS','ICICIBANK.NS','ITC.NS']  
for company in companies_list: 
    COMPANY = df.loc[:,(slice(None),company)] 
    COMPANY.columns = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']  
    COMPANY.to_csv(f'{company}.csv') 
```
